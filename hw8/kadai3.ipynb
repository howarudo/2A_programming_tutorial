```python
def rand_val():
    # generate random number between (0, 1)
    x = 1
    a = 1664525
    c = 1013904223
    m = 4294967296
    while True:
        xnext = (a * x + c) % m
        yield xnext / m
        x = xnext
```

```python
def random_walk(x0 = 0, p = 0.5):
    if p < 0 or p > 1:
        # raise error if p not in [0, 1]
        raise ValueError('p must be in [0, 1]')
    r = rand_val()
    x = x0
    x_list = [x0]
    for _ in range(1000):
        if next(r) < p:
            # with probability of p,
            x += 1
        else:
            # with probability of 1-p
            x -= 1
        x_list.append(x)
    return x_list
```

```python
import matplotlib.pyplot as plt
x0 = 0
p = 0.5
trail = random_walk(x0=x0, p=p)
plt.plot(trail)
```
