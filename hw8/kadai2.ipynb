```python
def logistic_recursion(X = 0.5, a = 3):
    if a < 0  or a > 4 or X < 0 or X > 1:
        # if a or X is not in the range specified raise ValueError
        raise ValueError('a  must be in [0, 4] and X must be in [0, 1]')
    while True:
        X_next = a * X * (1 - X) # recursion
        yield X_next
        X = X_next # renew the X
```

```python
a_list = []
X_list = []
order = 100
for a in range(1*order, 4*order):
    # range does not accept integer steps, so we use order
    a /= order
    g = logistic_recursion(a=a) # initiate first
    for i in range(500):
        if i > 50:
            X = next(g)
            # append a and X accordingly
            a_list.append(a)
            X_list.append(X)
```

```python
# appendがちゃんと行われたかを確認
assert len(a_list) == len(X_list)
```
