```python
def logistic_recursion(X = 0.5, a = 3):
    if a < 0  or a > 4 or X < 0 or X > 1:
        # if a or X is not in the range specified raise ValueError
        raise ValueError('a  must be in [0, 4] and X must be in [0, 1]')
    while True:
        X_next = a * X * (1 - X) # recursion
        yield X_next
        X = X_next # renew the X
```

```python
g = logistic_recursion(a=3.9, X=0.1)
traj = [next(g) for _ in range(100)]
import matplotlib.pyplot as plt
plt.plot(traj)
```
